```python
# Imports

import pandas as pd
#import pyspark as spark
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
```

### Preparing News Sentiment Dataset

```python
#News sentiment dataset
# df_news = pd.read_csv('cryptonews.csv')
df_news = pd.read_csv('cryptonews.csv')
df_news.columns = ['date','sentiment','source','subject','text','title','url']
df_news['date'] = df_news['date'].str[:10]
df_news.head()

#handling the Sentiment column, splitting the column into sentiment polarity and sentiment subjectivity
# Convert the strings in the 'sentiment' column to dictionaries
df_news['sentiment'] = df_news['sentiment'].apply(eval)

# Split the 'sentiment' column into three separate columns
df_news[['sentiment_class', 'sentiment_polarity', 'sentiment_subjectivity']] = pd.DataFrame(df_news['sentiment'].tolist())
df_news.drop(columns=['sentiment'], inplace=True)
df_news.drop(columns=['sentiment_class'], inplace=True)
display(df_news)
```

### Preparing BTC-USD Dataset

##### Data preprocessing: calculated and created price_change column, next_day_close and next_day_close_increased, before merging with the news sentiment dataset

```python

from copy import deepcopy
from sklearn.preprocessing import MinMaxScaler

#BTC-USD dataset
df_BTC = pd.read_csv('BTC-USD.csv')
df_BTC.columns = ['Date','Open','High','Low','Close','Adj CLose','Volume']


df_processed=deepcopy(df_BTC)
df_processed['price_change']=df_processed['Close']-df_processed['Close'].shift(-1)

# Creating price_change column that tracks the price change of the current day close and the previous day close
df_processed['next_day_close'] = df_processed['Close'].shift(-1)

#normalize price change and volume
scaler = MinMaxScaler()

#create new column to track whether the Close price increased the next date
df_processed['next_day_close_increased'] = (df_processed['Close'].shift(-1) < df_processed['Close']).astype(int)
BTC_processed_df = df_processed[:-1]


display(BTC_processed_df)
```

## Merging the datasets

```python
#standardisation of the columns
BTC_processed_df['Date'].str.strip()
df_news['date'].str.strip()

import pandas as pd

# Selecting columns from df_BTC
df_BTC_subset = BTC_processed_df[['Date', 'Close','price_change','next_day_close_increased']]

# Selecting column 'sentiment_polarity' from df_news
df_news_subset = df_news[['date', 'sentiment_polarity','sentiment_subjectivity']]
df_news_subset.columns = ['Date', 'sentiment_polarity','sentiment_subjectivity']  # renaming the columns to match

# Merging the two datasets on the 'Date' column
merged_df = pd.merge(df_BTC_subset, df_news_subset, on='Date', how='inner')

# Displaying the merged dataframe
display(merged_df)
```

#### Checking for class imbalances

```python
# merged_df.to_csv("merged_df.csv")
number_of_decrease = merged_df[merged_df['next_day_close_increased']==0].shape # to see the number of nextday decreases
number_of_increase = merged_df[merged_df['next_day_close_increased']==1].shape # to see the number of nextday increases

print(number_of_decrease)
print(number_of_increase)
```

### Train(90%), Validation(5%) and test(5%) split

```python
from math import floor

#train
train_end = floor(len(merged_df)*0.9)
train_df = merged_df[:train_end]
display(train_df)


#Valid
validation_end = floor(len(merged_df)*0.95)
valid_df = merged_df[train_end:validation_end]
display(valid_df)


#Test
test_df = merged_df[validation_end:]
display(test_df)
```

## Model 1 : Logistic Regression Model

#### Applying oversampling and undersampling due to imbalanced classes for next_day_close_increased column

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# Define the oversampling and undersampling techniques
oversampler = SMOTE(random_state=42)
undersampler = RandomUnderSampler(random_state=42)

# Define the pipeline with both oversampling and undersampling
pipeline = Pipeline([
    ('oversampling', oversampler),
    ('undersampling', undersampler)
])

```

### Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

features= ['sentiment_polarity', 'sentiment_subjectivity']
model = LogisticRegression()

#training# Apply the pipeline to training data
X_resampled, y_resampled = pipeline.fit_resample(train_df[features], train_df['next_day_close_increased'])

model.fit(X_resampled, y_resampled)

# Make predictions on the validation data
val_predictions = model.predict(valid_df[features])

all_predictions=model.predict(merged_df[features])

# Evaluate the model on the validation set
val_accuracy = accuracy_score(valid_df['next_day_close_increased'], val_predictions)
val_report = classification_report(valid_df['next_day_close_increased'], val_predictions)

# Display the results on the validation set
print("Validation Accuracy: {:.2f}".format(val_accuracy))
print("Validation Classification Report:\n", val_report)
```

### Evaluation

```python

test_predictions = model.predict(test_df[features])
# Evaluate the model on the test set
test_accuracy = accuracy_score(test_df['next_day_close_increased'], test_predictions)
test_report = classification_report(test_df['next_day_close_increased'], test_predictions)

# Display the results on the test set
print("\nTest Accuracy: {:.2f}".format(test_accuracy))
print("Test Classification Report:\n", test_report)
```

## Model 2: Simple NN

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
```

#### Model Architecture

```python
# neural network
#we are initialising with 3 linear layers and a droupout, Relu and sigmoid layer
class SimpleNN(nn.Module):
    def __init__(self, input_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.dropout = nn.Dropout(0.6)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.sigmoid(self.fc3(x))
        return x
```

#### Preparing features and outputs

```python
features = ['sentiment_polarity','sentiment_subjectivity']
X_train = train_df[features] #taking features that are defined above in the previous model
y_train = train_df['next_day_close_increased']


X_val = valid_df[features]
y_val = valid_df['next_day_close_increased']

X_test= test_df[features]
y_test = test_df['next_day_close_increased']

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val.values, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32)
```

#### Create DataLoader for training, validation, and testing sets

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
test_loader = DataLoader(test_dataset, batch_size=32)
```

#### Training the simple neural network

```python
# Initialize the neural network
input_size = X_train.shape[1]
model_SNN = SimpleNN(input_size)

# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.Adam(model_SNN.parameters(), lr=1e-2)
f1_scores = []
# Training loop
num_epochs = 15
for epoch in range(num_epochs):
    model_SNN.train()
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model_SNN(inputs)
        loss = criterion(outputs, labels.view(-1, 1))
        loss.backward()
        optimizer.step()

    # Validation
    model_SNN.eval()
    val_loss = 0.0
    correct = 0
    total = 0

    predictions = []
    true_labels = []

    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model_SNN(inputs)
            val_loss += criterion(outputs, labels.view(-1, 1)).item()

            # Calculate accuracy
            predicted = torch.round(outputs)
            total += labels.size(0)
            correct += (predicted == labels.view(-1, 1)).sum().item()

            predictions.extend(predicted.tolist())
            true_labels.extend(labels.tolist())

    avg_val_loss = val_loss / len(val_loader)
    val_accuracy = correct / total

     # Calculate F1 score
    f1 = f1_score(true_labels, predictions)
    # Store F1 score
    f1_scores.append(f1)

    print(f'Epoch [{epoch + 1}/{num_epochs}], Validation Loss: {avg_val_loss:.4f}')
    print(f'Epoch Accuracy: {val_accuracy:.4f}')

```

#### Testing

```python
# Evaluate the model on the test set
model_SNN.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = model_SNN(inputs)
        predicted = torch.round(outputs)
        total += labels.size(0)
        correct += (predicted == labels.view(-1, 1)).sum().item()

accuracy = correct / total
print(f'Test Accuracy: {accuracy:.4f}')
```

#### Classification Report of the SNN model results

```python

from sklearn.metrics import classification_report

# Convert the predictions and true labels to numpy arrays
predicted_array = np.array(predictions)
true_labels_array = np.array(true_labels)

# Convert binary predictions to integer format (0 or 1)
predicted_int = predicted_array.astype(int)
true_labels_int = true_labels_array.astype(int)

# Generate the classification report
report = classification_report(true_labels_int, predicted_int)

# Print the classification report
print(report)

#Graph
import matplotlib.pyplot as plt

# Assuming you have a list of validation losses for each epoch called 'validation_losses'
# Replace 'validation_losses' with your actual list of validation losses
```
