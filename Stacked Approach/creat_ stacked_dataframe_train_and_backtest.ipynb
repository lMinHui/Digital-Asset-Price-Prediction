```python

import pandas as pd
import numpy as np

df = pd.read_csv('final_combined_df.csv')
df
```

**Add in the outputs from BTC-only svm model**

```python
import pickle

with open('poly_svm_model_k5_BTC_only (3).pkl', 'rb') as f:
    trained_model = pickle.load(f)


selected_feats=['Open', 'High', 'Close', 'Adj Close', 'Volume', 'RSI_14', 'macd', 'macd_signal', 'BB_middle', 'BB_lower', 'Close-Open', 'High-Low', 'OBV', 'BB_height']
X_test = df[selected_feats]

predictions_BTC_only = trained_model.predict(X_test.values)
df['predictions_BTC_only']=predictions_BTC_only
df
```

**Add in outputs of tweets-only model**

```python
tweets_only_prediction = pd.read_csv('vinny_data-2.csv')
tweets_only_prediction=tweets_only_prediction.dropna()
tweets_only_prediction
```

```python
#since there are multiple outputs for same day, I will take the average
# Convert date column to datetime format (if it's not already)
tweets_only_prediction['Date'] = pd.to_datetime(tweets_only_prediction['Date'])

# Group by date and calculate the average
aggregated = tweets_only_prediction.groupby('Date')['Output'].mean().reset_index()
aggregated
```

```python
merged_df = pd.merge(df, aggregated, on='Date', how='inner')
merged_df
```

```python
merged_df=merged_df.rename(columns={'Output': 'prediction_Tweet_only'})
merged_df
```

**Add in outputs of news only model**

```python
news_only_prediction = pd.read_csv('vinny_news_data.csv')
news_only_prediction=news_only_prediction.dropna()
news_only_prediction
```

```python
news_only_prediction['Date'] = pd.to_datetime(news_only_prediction['Date'])

# Group by date and calculate the average
news_aggregated = news_only_prediction.groupby('Date')['Output'].mean().reset_index()
news_aggregated
```

```python
merged_df2 = pd.merge(merged_df, news_aggregated, on='Date', how='inner')
merged_df2
```

```python
merged_df2=merged_df2.rename(columns={'Output': 'prediction_News_only'})
merged_df2
```

```python
merged_df2.to_csv('combined_df_stacked.csv')
```

```python
models_only=merged_df2[['Date','next_day_close_increased','predictions_BTC_only','prediction_Tweet_only','prediction_News_only']]
models_only
```

```python
models_only.to_csv('models_only_stacked.csv')
```
